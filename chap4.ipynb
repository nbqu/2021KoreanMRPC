## 4.2 손실함수
### 오차제곱합 Sum of Squares Error, SSE

```python

def SSE(y, t):
    return np.sum((y-t)**2) * 0.5

y = np.array([0, 0, 1])
t = np.array([0, 0.3, 0.7])
SSE(y, t)
```


### 교차 엔트로피 오차 Cross Entropy Error, CEE
$E=-\sum_{k}t_{k}\log{y_k}$

정답일 때의 추정의 자연로그를 계산한다.<br>
즉, 교차 엔트로피 오차는 정답일 때의 출력이 전체 값을 정한다.

```python

def CEE(y, t):
    delta = 1e-7
    return -np.sum(t * np.log(y+delta))

t = np.eye(10)[2]
y = np.random.dirichlet(np.ones(10), 1) # 디리클레 분포 - 합이 1이 되는 무작위 분포
print(y)
print(CEE(y, t))
```


#### 손실함수를 사용하는 이유
'정확도'는 불연속적이다. 정확도는 가중치가 조금 변화했을 때 연속적으로 변하지 않는다.
